# Map parking data

```python
# import
import pandas as pd
import json
```

```python
# data
parking = pd.read_csv("data/Parking_Tickets.csv")
```

```python
print parking.shape
parking.head(5)
```

```python
with open('location.json', 'r') as infile:
    location = json.load(infile)
```

```python
parking["lon"] = 0
parking["lat"] = 0
```

```python
for r in range(parking.shape[0]):
    if r % 100 == 0:
        print r
    add = parking.irow(r)["LOCATION"]
    if location.has_key(add) and location[add] != []:
        res = location[add]
        parking.iloc[r,5] = res[0]
        parking.iloc[r,6] = res[1]
```

```python
parking.to_csv("data_geo/Parking_Ticket.csv",index=False)
```
